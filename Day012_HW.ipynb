```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text,"html5lib")

print(r.text[35100:])
```

### ① 取出今天所有的發文

```python
list1 = []
list2 = []
list3 = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    list1.append(d.find(class_='date').text)
    list2.append(d.find_all('a')[-1].text)
    
#print(list1)
#print(list2)
    

for date in list1:
    #print(type(date)) str
    #print(type(list1[0][0:10])) str
    
    if date[0:10] == list1[0][0:10]:
        list3.append(date)

#print(list3)
x = len(list3)
for today_title in list2[0:x+1]:
    print(today_title)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
res = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
#     tag = d.find(class_='tag').text
    tag = d.select('.tag')[0].text
    res.setdefault(tag, [])
    res[tag].append({
        'title': title,
        'date': date
    })
    
print(res)


'''
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find('em').text)
    print(d.find_all('a')[-1].text)
'''
```

### ③ 哪一個類別的文章最多

```python
from collections import Counter
list4 = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    list4.append(d.find('em').text)
    

word_counts = Counter(list4)
top_one = word_counts.most_common(1)
print('Ans:',top_one[0][0],'類別的文章最多')
```
